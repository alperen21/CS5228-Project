```python
import os
import pandas as pd
import numpy as np

df = pd.read_csv(os.path.join('/content/drive/MyDrive/', 'Project', 'cosine_cluster.csv'))
df.dropna(subset=['review_text'], inplace=True)
df.head()
```

```python
df['count'] = df.groupby('app_name')['app_name'].transform('count')
```

```python
# Check if there are any NaN values in the 'count' column
# Drop rows with NaN values in the DataFrame
df_cleaned = df.dropna()

df_cleaned

nan_in_count = df_cleaned['count'].isnull().any()

print(nan_in_count)
```

```python
import pandas as pd

# Assuming you have a DataFrame named df
# df = pd.DataFrame(data)

# Get all column names as an Index object
column_names = df.columns

# Print column names
print(column_names)

# If you want to convert it to a list
column_names_list = list(df.columns)

# Print the list of column names
sentiments = column_names_list[6:-2]
print(sentiments)
```

```python
df_x = df_cleaned[sentiments]
df_y = df_cleaned['count']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Example DataFrame 'df_features' contains features, and 'app_name_counts' is your target variable
# df_features, app_name_counts

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)

# Creating and fitting the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions
predictions = model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f"RMSE: {rmse}")
```

```python
# Coefficients of the model
weights = model.coef_

weights = zip(sentiments, weights)
# You can print out the weights for each feature
for weight in weights:
    print(weight)
```
